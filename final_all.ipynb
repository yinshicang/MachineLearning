```python
from selenium import webdriver #匯入
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from fake_useragent import UserAgent
import random
#import string
import time
```

```python
options = Options() 

ua = UserAgent()
userAgent = ua.random

options.add_argument(f'user-agent={userAgent}')
options.add_argument('--headless')  # 非本機執行，必須啟動Headless模式
options.add_argument('--disable-gpu') # 關閉GPU 避免某些系統或是網頁出錯

service = Service()
driver = webdriver.Chrome(service=service, options=options)
url = 'http://120.105.96.87/register_all.html' #網址可改
driver.get(url)
```

```python
from selenium.webdriver.support.ui import Select

myid=driver.find_element(By.NAME,'id') #尋找
myid.clear()
myid.send_keys("B09090088")#單一填選

myemail=driver.find_element(By.NAME,'email')
myemail.clear()
myemail.send_keys("B09090088@must.edu.tw")

passwd=driver.find_element(By.NAME,'pwd')
passwd.clear()
passwd.send_keys("B09090088")

sex=driver.find_element(By.ID,"f") #尋找id=f
#sex=driver.file_detector(By.XPATH,"//input[@value='f']")在沒有name的情況下
sex.click()#點選

interest=driver.find_element(By.ID,"tv")
interest.click()

#interest1=driver.find_element(By.ID,"movie") #網站設計movie是預設，再click就取消
#interest1.click()

#interest2=driver.find_element(By.ID,"bike")
#interest2.click()


county=Select(driver.find_element(By.NAME,'county'))
county.select_by_value("taoyuan")#select_by_value :Selenium下拉框處理的實現

driver.get_screenshot_as_file('data5.png')
mysubmit=WebDriverWait(driver,5).until(EC.presence_of_element_located((By.NAME,"submit")))
mysubmit.submit()
time.sleep(1)
driver.get_screenshot_as_file('data6.png')
driver.close()
```
