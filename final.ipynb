```python
from selenium import webdriver #匯入
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from fake_useragent import UserAgent
import random
#import string
import time
```

```python
options = Options() 

ua = UserAgent() #使用者使用的程式（軟體代理程式）
userAgent = ua.random

options.add_argument(f'user-agent={userAgent}')
options.add_argument('--headless')  # 非本機執行，必須啟動Headless模式
options.add_argument('--disable-gpu') # 關閉GPU 避免某些系統或是網頁出錯

service = Service()
driver = webdriver.Chrome(service=service, options=options)
url = 'http://120.105.96.87/register.html'#網址
driver.get(url)
```

### 註冊

```python
myid = driver.find_element(By.NAME, "id")#driver.find_element找尋name=("")
myid.clear()
myid.send_keys("B09090088") # .send_keys單一輸入

myemail = driver.find_element(By.NAME, "email")
myemail.clear()
myemail.send_keys("B09090088@must.edu.tw") # 請輸入你的email

passwd = driver.find_element(By.NAME, "pwd")
passwd.clear()
passwd.send_keys("B09090088")  # 請輸入你要設定的登入密碼

#selenium的截圖driver.get_screenshot_as_file()
driver.get_screenshot_as_file('data1.png') # 註冊畫面
mysubmit = WebDriverWait(driver, 5).until(EC.presence_of_element_located((By.NAME, "submit")))#WebDriverWait(等待頁面加載完成.直到)EC.presence_of_element_located(尋找目前出現在畫面上的元素)
mysubmit.submit()
time.sleep(1)
driver.get_screenshot_as_file('data2.png') # 註冊結果畫面 
driver.close()
```

### 登入

```python
service1 = Service()
driver1 = webdriver.Chrome(service=service1, options=options)#webdriver.Chrome 啟動 Chrome 瀏覽器的 WebDriver
url = 'http://120.105.96.87/index.html'
driver1.get(url)
myid = driver1.find_element(By.NAME, "id")#尋找name=id
myid.clear()
myid.send_keys("B09090088") # 請輸入你的學號
passwd = driver1.find_element(By.NAME, "pwd")
passwd.clear()
passwd.send_keys("B09090088")  # 請輸入你設定的登入密碼
driver1.get_screenshot_as_file('data3.png') # 登入畫面
mysubmit = WebDriverWait(driver1, 5).until(EC.presence_of_element_located((By.NAME, "submit")))#登入()
mysubmit.submit()
time.sleep(1)
driver1.get_screenshot_as_file('data4.png') # 登入結果畫面
driver1.close()
```
